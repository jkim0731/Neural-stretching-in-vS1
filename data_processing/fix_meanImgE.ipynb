```python
mouse=52
base_dir = Path('D:/')
mouse_dir = base_dir / f'{mouse:03}'
for plane in range(1,9):
    plane_dir = mouse_dir / f'plane_{plane}'
    session_dir_names = []
    for p in plane_dir.iterdir():
        if p.is_dir():
            if p.name[0].isnumeric:
                session_dir_names.append(p.name)
    for sn in session_dir_names:
        ops_fn = plane_dir / sn / 'plane0' / 'ops.npy'
        ops = np.load(ops_fn, allow_pickle=True).item()
        if 'meanImgE' not in ops.keys():
            ops = extraction.enhanced_mean_image(ops)
            np.save(f'{base_dir}{mouse:03}/plane_{plane}/{sn}/plane0/ops.npy', ops)
            print(f'ops saved for {mouse:03} {sn} plane {plane}')

```
